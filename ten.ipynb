```python
# Another way to do it without clearing the Axis

import pandas as pd # type: ignore
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()

def animate(i):
    try:
        data = pd.read_csv('data6.csv')
        print(f"Frame {i} | Rows: {len(data)}")

        x = data['x_value']
        y1 = data['total_1']
        y2 = data['total_2']

        ax.clear()

        ax.plot(x, y1, label='Channel 1')
        ax.plot(x, y2, label='Channel 2')

        ax.legend(loc='upper left')
        ax.set_title('Live Plot from CSV')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')

    except Exception as e:
        print(f"Error at frame {i}: {e}")

anim = FuncAnimation(fig, animate, interval=1000, cache_frame_data=True)

anim.save('anim.txt')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import time

plt.ion()  # Turn on interactive mode
fig, ax = plt.subplots()
x, y = [], []
line, = ax.plot(x, y)

for i in range(100):
    x.append(i)
    y.append(np.random.rand())
    line.set_xdata(x)
    line.set_ydata(y)
    ax.relim()
    ax.autoscale_view()
    plt.draw()
    plt.pause(0.1)  # Pause to update plot
```
